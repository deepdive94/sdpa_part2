## Customer Class

```python


class customer: # First class, that is "customer"

    def final_booking(car_type_ip,rental_list,hatchback,sedan,suv):  # Function initiation
                
        if "hatchback" in car_type_ip:  
            if car_name_ip in hatchback:
                hatchback.remove(car_name_ip)
                rental_list.append(car_name_ip)

        elif "sedan" in car_type_ip:
            if car_name_ip in sedan:
                sedan.remove(car_name_ip)
                rental_list.append(car_name_ip) 

        elif "suv" in car_type_ip:
            if car_name_ip in suv:
                suv.remove(car_name_ip)
                rental_list.append(car_name_ip) 
                
        return hatchback,sedan,suv,rental_list
        
    def bill_calc(days,price):  # Function to calculate the bill
        total = int(days)*int(price)
        return total     
 
```

## Rental Class

```python
   
class rental:     # rental class definition
    def check_rent_list(rental_list): # fucntion declaration 
#         print("Rental cars list in RENTAL CLASS: ", rental_list)
        ret_car_nm = input("Choose the car you want to return")
        return_index = rental_list.index(ret_car_nm)
        rental_list.remove(ret_car_nm)
        return rental_list, return_index,ret_car_nm
    
    def bill(days=None,price=None):
        total_bill = customer.bill_calc(days,price)
        return total_bill
    
```

### Dictionary Creation & Definition

```python
import json

type_list = ["hatchback","sedan","suv"]  # list with car types
less_than_week = ["30","50","100"]  # list prices of cars when hired for less than 7 days 
more_than_week = ["25","40","90"]  # list prices of cars when hired for more than 7 days


data_dict = { "data":
                 {
                    "rental_car": [],
                    "customers_list": [],
                    "less_than_week":[],
                    "rental_list":[],
                    "total_list":[],
                    "car_type":
                               { 
                                "type_order":["hatchback","sedan","suv"],
                                "hatchback":["h1","h2","h3","h4"],
                                "sedan":["s1","s2","s3"],
                                "suv" : ["su1","su2","su3"]
                               }
                }
            }
```
